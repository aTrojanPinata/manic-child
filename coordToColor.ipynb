```
%matplotlib inline
import numpy as np
import pylab
from mpl_toolkits.mplot3d import Axes3D # for the 3D plots we'll make below

```

```
def coordToColor(r,g,b):
    #if r.shape!=g.shape!=b.shape:
        #print('values have different lengths')
        #break
    rr=np.zeros_like(r)
   
    for k in range(len(r)):
        if r[k] == 0:
            rr[k]=0
        else:
            rr[k]=(abs(r[k]/np.max(abs(r))))
       
    
    gg=np.zeros_like(r)
    
    for k in range(len(r)):
        if g[k] == 0: 
            gg[k]=0
        else:
            gg[k]=(abs(g[k]/np.max(abs(g))))
            
            
    bb=np.zeros_like(r)
    
    for k in range(len(r)):
        if b[k] ==0:
            bb[k]=0
        else:
            bb[k]=(abs(b[k]/np.max(abs(b))))

    colorcolumns=np.append([rr,gg],[bb],axis=0)
    colorrows=np.swapaxes(colorcolumns,0,1)
    return(colorrows)
```

```
def coordToColorNeg(r,g,b):
    r=r-np.min(r)
    g=g-np.min(g)
    b=b-np.min(b)
    
    rr=np.zeros_like(r)
   
    for k in range(len(r)):
        if r[k] == 0:
            rr[k]=0
        else:
            rr[k]=(abs(r[k]/np.max(abs(r))))
       
    
    gg=np.zeros_like(r)
    
    for k in range(len(r)):
        if g[k] == 0: 
            gg[k]=0
        else:
            gg[k]=(abs(g[k]/np.max(abs(g))))
            
            
    bb=np.zeros_like(r)
    
    for k in range(len(r)):
        if b[k] ==0:
            bb[k]=0
        else:
            bb[k]=(abs(b[k]/np.max(abs(b))))

    colorcolumns=np.append([rr,gg],[bb],axis=0)
    colorrows=np.swapaxes(colorcolumns,0,1)
    return(colorrows)
```

```
#Copied Import Statements

fig = pylab.figure(figsize=(4,4))
ax = Axes3D(fig)
k=1000
line=np.linspace(0,16*np.pi, k)
zoos=np.linspace(0,0,k)
zoo=zoos+1


x=np.e**zoos
y=np.cos(line)
z=np.sin(line)

ax.scatter(zoo,y,z, c= coordToColor(x,z,y), marker = 'o') 
```
